# Data cleaning

```python

import pandas as pd
df=pd.read_csv("C:/Users/punit/OneDrive/Documents/Job_market_intelligence_project/data/postings.csv")
df.shape
```

```python
drop_cols = [
    "sponsored",
    "fips",
    "zip_code",
    "job_posting_url",
    "application_url",
    "posting_domain",
    "skills_desc",
    "closed_time",
    "med_salary"
]

df = df.drop(columns=drop_cols)
```

### cleaning normalized salary

```python
df["normalized_salary"].isnull().sum()
```

```python
df["normalized_salary"].describe()
```

```python
df["normalized_salary"].sort_values(ascending=False).head(20)
```

```python
df["normalized_salary"].sort_values().head(20)
```

```python
## inter quartile range method
q1 = df["normalized_salary"].quantile(0.25)
q3= df["normalized_salary"].quantile(0.75)
IQR=q3-q1

print("q1:",q1)
print("q2:",q3)
print("IQR:", IQR)

print("Lower bound:", (q1-(1.5*IQR)))
print("Upper bound:", (q3+(1.5*IQR)))
```

#### =>IQR analysis identified potential outliers. Since salary cannot be negative and some high-paying professions legitimately earn above the statistical upper bound, domain-specific thresholds of 15,000 and 600,000 were used instead.

```python
df[df["normalized_salary"]<15000].shape
```

```python
df[df["normalized_salary"] > 600000].shape
```

```python
before_rows = len(df)
print("before_rows:" ,before_rows)
df = df[(df["normalized_salary"].isnull())
    |
    ((df["normalized_salary"] > 15000) & (df["normalized_salary"] < 600000))
]
after_rows=len(df)
print("after_rows:" , after_rows)
print("% of rows removed:",((before_rows-after_rows)/before_rows)*100)
```

### cleaning formatted_work_exp 

```python
df["formatted_experience_level"]=df["formatted_experience_level"].fillna("Not specified")
df["formatted_experience_level"].value_counts()
```

## cleaning locations

```python
df["location"].value_counts().head(100)
```

```python
df["clean_location"]=df["location"].str.split(",").str[0]
df["clean_location"].value_counts().head(100)
```

```python
df.isnull().sum()
```

```python
## dropping unnecessary columns
df=df.drop(columns=["max_salary","pay_period","currency","compensation_type","min_salary","company_id"])
```

## Correlation heatmap:

```python
df=pd.read_csv("C:/Users/punit/OneDrive/Documents/Job_market_intelligence_project/data/cleaned_postings.csv")
df.head()
```

```python
df.select_dtypes(include="number").columns
```

```python
corr=df.select_dtypes(include="number").corr()
corr
```

```python
import seaborn as sns
sns.heatmap(corr,annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

The strongest relationship in the dataset is between views and applies, with a correlation of 0.48. This indicates a moderate positive relationship, meaning jobs that receive more views tend to receive more applications.

Normalized salary shows almost no correlation with views (0.045) or applies (0.032), suggesting that higher-paying jobs do not necessarily attract significantly more views or applications.

Job ID has virtually no correlation with any variable, which is expected since it is simply an identifier and does not carry meaningful business information.
